```python
import gradio as gr
import os
import random

# Simple response function for demo purposes
# In a real application, you would implement an actual NLP model or API call
def get_response(user_input):
    """Generate response based on user input"""
    # Sample banking responses for demonstration
    banking_responses = {
        "transfer": "To transfer money between accounts, log in to VCB Digibank, select 'Transfer', choose the source and destination accounts, enter amount, and confirm the transaction.",
        "balance": "You can check your account balance by logging into VCB Digibank, using Vietcombank ATMs, or contacting our customer service at 1900 54 54 13.",
        "card": "Vietcombank offers various card options including debit, credit, and prepaid cards. Each has different features and benefits tailored to your needs.",
        "loan": "Vietcombank provides various loan products including home loans, car loans, and personal loans with competitive interest rates.",
        "account": "To open a new account, please visit your nearest Vietcombank branch with your ID card/passport and proof of address.",
        "fee": "Bank fees vary depending on the service. You can find our complete fee schedule on the Vietcombank website.",
        "branch": "You can find your nearest Vietcombank branch by using the branch locator on our website or mobile app.",
        "interest": "Current savings interest rates range from 3.8% to 5.6% depending on term length and deposit amount."
    }

    # Simple keyword matching (in a real app, use NLP for better understanding)
    for keyword, response in banking_responses.items():
        if keyword.lower() in user_input.lower():
            return response

    # Default response if no keywords match
    return "Thank you for your question. For specific banking inquiries, please contact our customer service at 1900 54 54 13 or visit your nearest Vietcombank branch."

# Chat history function to maintain conversation state
def respond(message, chat_history):
    """Process user message and update chat history"""
    bot_message = get_response(message)
    chat_history.append((message, bot_message))
    return "", chat_history

# Custom theme with Vietcombank branding colors
vcb_theme = gr.themes.Base(
    primary_hue="green",
    secondary_hue="green",
    neutral_hue="gray"
).set(
    button_primary_background_fill="#007A33",
    button_primary_background_fill_hover="#004B23",
    button_primary_text_color="white",
    button_secondary_background_fill="#E3F1EB",
    button_secondary_background_fill_hover="#C7E4D5",
    button_secondary_text_color="#007A33",
    container_radius="10px",
    input_background_fill="#FFFFFF",
    input_border_color="#007A33",
    input_border_width="2px",
    input_radius="10px",
    block_title_text_weight="600",
    block_label_text_color="#0a4d2e"
)

# Custom CSS for additional styling
custom_css = """
.gradio-container {
    font-family: 'Segoe UI', Arial, sans-serif;
    max-width: 900px;
    margin: auto;
}

#header-area {
    text-align: center;
    margin-bottom: 20px;
}

#logo-image {
    max-height: 60px;
    margin-bottom: 10px;
}

footer {
    margin-top: 20px;
    text-align: center;
    font-size: 12px;
    color: #666;
}

.feedback-buttons {
    display: flex;
    justify-content: center;
    gap: 10px;
}

.bold-text {
    font-weight: bold;
    color: #007A33;
}
"""

# Sample frequently asked questions
sample_questions = [
    "How do I transfer money to another bank?",
    "What are the current interest rates for savings?",
    "How can I check my account balance?",
    "What documents do I need to open a new account?",
    "How do I apply for a credit card?",
    "What are the fees for international transfers?"
]

# Create the Gradio interface
with gr.Blocks(theme=vcb_theme, css=custom_css) as demo:
    # Header with logo and title
    with gr.Row(elem_id="header-area"):
        with gr.Column():
            gr.HTML("""
                <img id="logo-image" src="https://upload.wikimedia.org/wikipedia/vi/7/7c/Vietcombank_Logo.png" alt="Vietcombank Logo">
                <h1 style="color: #007A33; margin: 5px 0 20px 0;">Banking Assistant</h1>
                <p>Welcome to Vietcombank Virtual Assistant. How can I help you today?</p>
            """)

    # Chatbot interface
    chatbot = gr.Chatbot(
        label="Conversation",
        bubble_full_width=False,
        show_copy_button=True,
        height=400
    )

    # Input area with clear button and examples
    with gr.Row():
        msg = gr.Textbox(
            label="Your Message",
            placeholder="Type your banking question here...",
            scale=9
        )
        submit_btn = gr.Button("Send", variant="primary", scale=1)

    # Example questions for quick access
    gr.Examples(
        examples=sample_questions,
        inputs=msg,
        label="Example Questions"
    )

    # Clear button for chat history
    clear = gr.Button("Clear Conversation")

    # Feedback buttons
    with gr.Row(elem_id="feedback-buttons"):
        gr.HTML("""
            <div class="feedback-buttons">
                <button onclick="alert('Thank you for your feedback!')" style="background-color: #007A33; color: white; border: none; padding: 8px 15px; border-radius: 5px; cursor: pointer;">ğŸ‘ Helpful</button>
                <button onclick="alert('We\'ll improve our responses. Thanks!')" style="background-color: white; color: #007A33; border: 1px solid #007A33; padding: 8px 15px; border-radius: 5px; cursor: pointer;">ğŸ‘ Not Helpful</button>
            </div>
        """)

    # Footer with disclaimer
    gr.HTML("""
        <footer>
            <p>Â© 2025 Vietcombank. This is a demo assistant. For official banking services, please visit <span class="bold-text">vietcombank.com.vn</span> or call <span class="bold-text">1900 54 54 13</span>.</p>
        </footer>
    """)

    # Set up event handlers
    submit_btn.click(respond, [msg, chatbot], [msg, chatbot])
    msg.submit(respond, [msg, chatbot], [msg, chatbot])
    clear.click(lambda: None, None, chatbot, queue=False)

# Launch the app
if __name__ == "__main__":
    demo.launch(share=True)  # Set share=False in production
```
